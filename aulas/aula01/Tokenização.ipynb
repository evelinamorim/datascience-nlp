# Tokenização com o NLTK: Natural Language Toolkit

```python
# certifique-se que você tem o corpus tentando fazer o download. Chamamos de corpus um conjunto de textos 
# escritos com um objetivo em particular. Em NLP cada conjunto de dados pode se chamar corpus e mais de um conjunto
# se chama corpora
import nltk
nltk.download("machado")
```

```python
from nltk.corpus import machado # importando o corpus machado de livros do Machado de Assis
```

```python
print(machado.readme()) # algumas informações sobre esse corpus
```

```python
print(machado.fileids()) # arquivos deste corpus
```

```python
conto01 = machado.raw(machado.fileids()[0]) # vamos pegar o conteudo do texto "cru" deste primeiro texto

# você pode imprimir o conteudo de conto01 e se certificar do que a variável contém
```

```python
from nltk import word_tokenize # importando a função que tokeniza um texto "cru"
```

```python
tokens_conto01 = word_tokenize(conto01)
```

```python
print(tokens_conto01)
```

# Tokenização com o Spacy

Certifique-se que você tem em seu computador o modelo português do spacy
Caso não tenha, basta executar "python -m spacy download pt_core_news_sm" na sua linha de comando

```python
import spacy
```

```python
nlp = spacy.load("pt_core_news_sm")
```

```python
# realiza a tokenização e outras tarefas de nlp, neste exemplo vamos mostrar o resultado apenas da tokenização
doc_conto01 = nlp(conto01) 
```

```python
for token in doc_conto01:
    print(token.text)
```
