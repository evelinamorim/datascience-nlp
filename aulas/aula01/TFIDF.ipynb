```python
from gensim.utils import simple_preprocess
from nltk.corpus import machado
```

```python
corpus_machado = []
for f_id in machado.fileids():
    corpus_machado.append(simple_preprocess(machado.raw(f_id)))
```

```python
count_capitu = {}
count_doutor = {}

for idx, doc in enumerate(corpus_machado):
    
    if "capitu" in doc:
        print(idx)
        if "doc" in count_capitu:
            count_capitu["doc"] += 1
        else:
            count_capitu["doc"] = 1
            
    if "doutor" in doc:
        if "doc" in count_doutor:
            count_doutor["doc"] += 1
        else:
            count_doutor["doc"] = 1
        
    for token in doc:
        
        if token == "capitu":
            if "term" in count_capitu:
                count_capitu["term"] += 1
            else:
                count_capitu["term"] = 1
                
        if token == "doutor":
            if "term" in count_doutor:
                count_doutor["term"] += 1
            else:
                count_doutor["term"] = 1
 
```

```python
from gensim.models import TfidfModel
from gensim.corpora import Dictionary
```

```python
dct = Dictionary(corpus_machado)
```

```python
# converte o corpus para o formato de saco de palavras
corpus_machado_bow = [dct.doc2bow(line) for line in corpus_machado]  
```

```python
model = TfidfModel(corpus_machado_bow) # computa para cada termo, a fórmula tf-idf
```
